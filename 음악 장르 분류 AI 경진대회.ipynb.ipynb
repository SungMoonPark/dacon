```python
X = train.drop(['genre'], axis= 1)
y= train['genre']
```

```python
import lightgbm as lgb
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
test.drop(['ID'], axis=1, inplace = True)
```

```python
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
```

```python
X_train, X_val, y_train, y_val = train_test_split(X,y, test_size = 0.2)
```

```python
clf = lgb.LGBMClassifier()
clf.fit(X_train,y_train)
```

```python
params = {'max_depth' : [10,13,15,17,19], 'min_child_weight' : [1,2], 'learning_rate':[0.01, 0.05, 0.1]}
```

```python
from sklearn.metrics import f1_score, make_scorer
```

```python
f1 = make_scorer(f1_score , average='macro')
```

```python
grid_cv = GridSearchCV(clf, param_grid = params, cv = 3, scoring = f1)
grid_cv.fit(X_train, y_train)
```

```python
grid_cv.best_params_
```

```python
clf = lgb.LGBMClassifier(learning_rate = 0.05, max_depth = 15, min_child_weight = 2)
clf.fit(X_train,y_train)
```

```python
prd = clf.predict(X_val)
print(classification_report(y_val,  prd))
```

```python
pre = clf.predict(test)
```
